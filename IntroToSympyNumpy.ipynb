```python
from sympy import*
from numpy import*
from matplotlib.pyplot import*

ar=arange(-15,15)
x = symbols('x')
f = x**2
f = lambdify(x,f)
print(ar)
print(f(ar))
plot(ar, f(ar))
xlabel('x axis')
ylabel('y axis')
title('Function f')
show()
```

```python
from matplotlib.pyplot import*
f = 2*x + x**4
f = lambdify(x,f)
plot(ar,f(ar))
xlabel('x-axis')
ylabel('y-axis')
title('Function 2*x + x**4')
show()
```
